```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace("05-Oct","5-10")
df.head()
```

```python
ax = sns.countplot(x='Gender',data=df, hue='Gender')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analyse that  the number of females in the data is more than the number of males

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
gb
```

```python
sns.heatmap(gb,annot = True)
plt.title("Relationship between parents education and students score")
plt.show()
```

From above chat we have concluded
that the education of parents have a good impact on ther scores

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
gb1
```

```python
sns.heatmap(gb1,annot = True)
plt.title("Relationship between parents maritalStatus and students score")
plt.show()
```

From above chat we have concluded that the marital Staus of parents have a negligible impact on ther scores

```python
sns.boxplot(data=df,x="MathScore")
plt.show()
```

```python
df['EthnicGroup'].unique()
```

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ['group A','group B','group C','group D','group E']
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]

plt.pie(mlist,labels=l,autopct='%1.2f%%')
plt.title("Ethnic Group Distribution")
plt.show()
```
